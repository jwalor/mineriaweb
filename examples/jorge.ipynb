### Extraccion de Datos Basica

````python
#%time
tweets = tweepy.Cursor(api.search, q=query, tweet_mode='extended', lang="es", since=date_since).items(2000)
````

````python
# Recolectando los datos del tweet
all_tweets = [[tweet.user.id, tweet.user.screen_name, tweet.user.name, tweet.user.created_at, tweet.full_text] for tweet in tweets]
````

### Extraccion de Datos con Timer

````python
%time
import time

all_tweets2 = []
c = 0

for status in tweepy.Cursor(api.search, 
                           q=query,
                           count=10000,
                           rpp=10000,
                           monitor_rate_limit=True,
                           wait_on_rate_limit=True,
                           wait_on_rate_limit_notify=True,
                           retry_count=5,
                           retry_delay=5
                           ).items():
    try:
        all_tweets2.append([status.id, status.user.id, status.user.screen_name, status.user.name, status.user.created_at, status.text])
    
    except Exception as e:
        print(e)
        
        c += 1
        
        if c % 100 == 0:
            time.sleep(5)
````
